```python
yellow_total = yellow_trips.shape[0]
yellow_credit_card = yellow_trips[(yellow_trips.payment_type == 1)]
ycc_total = yellow_credit_card.shape[0]
ycc_tipped = yellow_credit_card[(yellow_credit_card.tip_amount > 0.01)].shape[0]
ycc_ratio = round(ycc_total/yellow_total, 3)
print(f"Y_total: {yellow_total}, YCC_total: {ycc_total}, YCC_tipped: {ycc_tipped}")
print(f"ycc_tipped/ycc_total: {round(ycc_tipped/ycc_total*100, 3)}% of all Credit Card trips were tipped.")
print(f"ycc/y_total: {ycc_ratio*100}% of all trips were payed with Credit Card.")
```

```python
yellow_total = yellow_trips.shape[0]
yellow_cash = yellow_trips[(yellow_trips.payment_type == 2)]
ycash_total = yellow_cash.shape[0]
ycash_tipped = yellow_cash[(yellow_cash.tip_amount > 0.01)].shape[0]
ycash_ratio = round(ycash_total/yellow_total, 3)
print(f"Y_total: {yellow_total}, YCASH_total: {ycash_total}, YCC_tipped: {ycash_tipped}")
print(f"ycash_tipped/ycash_total: {round(ycash_tipped/ycc_total*100, 3)}% of all Cash trips were tipped.")
print(f"ycash/y_total: {ycash_ratio*100}% of all trips were payed with Cash.")
```

```python
ycc_subset = yellow_credit_card.sample(int(1e6*ycc_ratio), random_state=87)
yc_subset = yellow_cash.sample(int(1e6*ycash_ratio), random_state=87)
```
